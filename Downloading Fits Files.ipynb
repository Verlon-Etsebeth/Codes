```python
import numpy as np

# Set up matplotlib and use a nicer set of plot parameters
%config InlineBackend.rc = {}
import matplotlib

import matplotlib.pyplot as plt
%matplotlib inline

from astropy.utils.data import download_file
from astropy.io import fits
```

```python
## Image tractor

#image_file_tractor = download_file('tractor-0001m002', cache=True )
hdu_list = fits.open('tractor-0001m002.fits', cache=True )
hdu_list.info()
image_data_tractor = fits.getdata('tractor-0001m002.fits',0)
print(type(image_data_tractor))
print(image_data_tractor.shape)

#outfile = 'legacysurvey-0001m002-image-tractor.fits'

#hdu = fits.PrimaryHDU(image_data_tractor)
#hdu.writeto(outfile, overwrite=True)
```

```python
## Image G band

image_file_g_band = download_file('https://portal.nersc.gov/project/cosmo/data/legacysurvey/dr8/south/coadd/000/0001m002/legacysurvey-0001m002-image-g.fits.fz', cache=True )
hdu_list = fits.open(image_file_g_band)
hdu_list.info()
image_data_g_band = fits.getdata(image_file_g_band)
print(type(image_data_g_band))
print(image_data_g_band.shape)
plt.imshow(image_data_g_band, cmap='gray')
plt.colorbar()

outfile = 'legacysurvey-0001m002-image-g.fits'

hdu = fits.PrimaryHDU(image_data_g_band)
hdu.writeto(outfile, overwrite=True)
```
